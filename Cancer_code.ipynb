```python
Q1=np.percentile(df['radius_mean'],25) #lb
Q3=np.percentile(df['radius_mean'],75) #ub
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
c1=df['radius_mean']<lb
c2=df['radius_mean']>ub
c=c1|c2
outlier_df=df[c]
outlier_df
```

**Feature selection**

```python
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                              random_state=42,
                                              test_size=0.30)
```

**Model**

```python
#Logistic Regression
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)

```

**Hypertunning Model**

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
dt_params=[keys for keys in dtree.get_params()]

param_grid = {
    'criterion":['gini','entropy'], # 2
    'max_depth': [3, 5, 10],     # 3
    'max_features': ['sqrt', 'log2'], # 2
    'random_state': [0, 42] # 2
    
}

grid_search = GridSearchCV(dtree,  # Base model
                           param_grid, # params 
                           scoring='accuracy', # metric
                           cv=5,
                          verbose=True)

grid_search.fit(x_train,y_train)

best_params=grid_search.best_params_
best_score=grid_search.best_score_
print("best params:",best_params)
print("best train accuracy:",best_score)

df
```

**Model Prediction**

```python
y_pred_log=dtree.predict(x_test)

cancer_test_df=pd.DataFrame(zip(ids,y_pred_log),
                          columns=['id','diagnosis'])
```
